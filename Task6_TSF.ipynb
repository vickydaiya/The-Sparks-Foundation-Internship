### **Importing libraries and functions**

```
from google.colab import files
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image  
```

### **Uploading dataset**

```
files.upload()
```

### **Reading dataset**

```
iris_data = pd.read_csv('Iris.csv')
iris_data.head()
```

### **Cleaning data**

```
data = iris_data.drop('Id',axis='columns')
predictor_data = data.drop('Species',axis='columns') 
response_data = data['Species']
```

### **Creating training and test sets**

```
predictor_train_data, predictor_test_data, response_train_data, response_test_data = train_test_split(predictor_data, response_data, test_size = 0.2)
```

### **Creating a decision tree on training data**

```
dt = DecisionTreeClassifier()

dt = dt.fit(predictor_train_data,response_train_data)

predictions = dt.predict(predictor_test_data)
```

### **Model evaluation**

```
print("Accuracy:",metrics.accuracy_score(response_test_data, predictions))
```

### **Visualizing the decision tree**

```
dot_data = StringIO()
export_graphviz(dt, out_file=dot_data, feature_names=predictor_data.columns,  
                filled=True, rounded=True,
                special_characters=True,class_names=dt.classes_)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
